```python
features_high_corr = []
for i in KDD_corr_mat:
    if KDD_corr_mat['class'][i] > 0.2 or KDD_corr_mat['class'][i] < -0.2:
        features_high_corr.append(i)
features_high_corr
```

```python
features_dropped = []
for i in KDD_corr_mat.columns:
    if i not in features_high_corr:
        features_dropped.append(i)
features_dropped
```

```python
collinear_features = {}
for i in KDD_corr_mat_final:
    collinear_features[i] = []
    for j in KDD_corr_mat_final:
        if i != j and (KDD_corr_mat_final[i][j] > 0.8 or KDD_corr_mat_final[i][j] < -0.8):
            collinear_features[i].append(j)
collinear_features
```

```python
dropped_collinear_features = [ 
    'srv_serror_rate',
    'dst_host_serror_rate',
    'dst_host_srv_serror_rate',
    'dst_host_same_srv_rate',
    'dst_host_rerror_rate',
    'srv_rerror_rate',
    'dst_host_srv_rerror_rate'
    ]
```

```python
X = KDD_data.iloc[:, :-1]
Y = KDD_data.iloc[:, -1]
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV
from sklearn.pipeline import make_pipeline
```

```python
model_params = {
    'Naive_Bayes' : {
      'model' : GaussianNB(),
      'params' : {
          
      }  
    },
    'random_forest' : {
        'model' : RandomForestClassifier(),
        'params' : {
            'randomforestclassifier__n_estimators' : [100, 110, 120, 150, 200]
        }
    },
    'logistic_regression': {
        'model' : LogisticRegression(),
        'params' : {
            'logisticregression__solver' : [ 'newton-cg', 'sag', 'saga'],
            'logisticregression__C' : [1, 10, 20]
        }
    },
    'KNN' : {
      'model' : KNeighborsClassifier(),
      'params' : { 
          'kneighborsclassifier__n_neighbors' : [3, 4, 5]
      } 
    }
}
```

```python
scores = []
best_esmti = {}

for algo, mp in model_params.items():
    pipe = make_pipeline(mp['model'])
    clf = GridSearchCV(pipe, mp['params'], cv=5, return_train_score=False)
    clf.fit(X, Y)
    scores.append({
        'model' : algo,
        'best_score' : clf.best_score_,
        'best_params' : clf.best_params_
    })
    best_esmti[algo] = clf.best_estimator_
```

```python
df = pd.DataFrame(scores)
df
```

```python
classifier_df = df.iloc[:, :-1]
classifier_df.columns = ['Alogrithm', 'Accuracy']
classifier_df = classifier_df.sort_values(by='Accuracy', ascending=False)
```

```python
print("PARAMETER OPTIMISED ALGORITHM MODELS:")
classifier_df.reset_index(drop='index')
```

```python
matrices = []
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
for i in best_esmti:
    y_temp_pred = best_esmti[i].predict(X)
    matrices.append(confusion_matrix(Y, y_temp_pred))
    print(i+"\n",classification_report(Y, y_temp_pred))
```
